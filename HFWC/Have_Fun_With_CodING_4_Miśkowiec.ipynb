````python
def hfwc1(input):
    output = {
        'bool' : [],
        'float' : [],
        'int' : [],
        'str' : []
        }
    
    for element in input:
        for key, value in output.items(): 
            if type(element) == bool and key == 'bool':
                value.append(element)
            elif type(element) == float and key == 'float':
                value.append(element)
            elif type(element) == int and key == 'int':
                value.append(element)
            elif type(element) == str and key == 'str':
                value.append(element)            
    return output
    
print(hfwc1(input))
````

Potrzeba napisać oprogramowanie do obsługi kantoru. Kantor ma tablicę z cenami poszczególnych walut ```price_table```, oraz określoną ilość gotówki w kasie w różnych walutach ```cash_register_data```. Praca kantoru polega na wykonywaniu transakcji kupna lub sprzedaży walut. Transakcja może się udać, lub nie być możliwa do wykonania z powodu zbyt małej ilości gotówki w kasie określonego typu (np. klient chce kupić 100USD a kantor ma tylko 80USD w kasie).

Przykłady transakcji:
```python
Transaction('BUY', 2800, 'CHF')
# oznacza, że KANTOR KUPUJE 2800 Franków i wypłaca złotówki

Transaction('SELL', 1200, 'USD')
# oznacza, że KANTOR SPRZEDAJE 1200 Dolarów i przyjmuje złotówki
```

Dane wejściowe:
- **price_table** - tablica cen kupna / sprzedaży - pierwsza w tupli cena po której kantor kupuje, druga - po której sprzedaje
- **cash_register_data** - zawartość kasy - ilości poszczególnych walut w kasie
- **transactions** - lista transakcji do wykonania

- podjąć próbę wykonania transakcji z listy po kolei

# 3. ToDoList (20 pkt.)

Zadanie polega na  napisaniu prostej aplikacji przechowującej listę rzeczy do zrobienia. 

Aplikacja ma umożliwiać:

* dodawanie nowych zadań `metoda add_task`, 
* oznaczanie zadań jako wykonanych `metoda finish`
* zwracanie na ekran listy zadań z ich statusem (wykonane / niewykonane) `metoda print_list`. Zadanie oznaczone jako wykonane otrzymuje znak `[x]`

Aplikacja ma działać w taki sposób, aby wykonanie kodu:
```python
mylist = ToDoList('My sample ToDoList')
for i in range(1, 11):
    task = Task(f'Simple task #{i}')
    if i % 2:
        task.finish()
    mylist.add_task(task)
mylist.print_list()
```

zwróciło (wypisało na ekran) dokładnie to co poniżej:
```
My sample ToDoList
###
[x] Simple task #1
[ ] Simple task #2
[x] Simple task #3
[ ] Simple task #4
[x] Simple task #5
[ ] Simple task #6
[x] Simple task #7
[ ] Simple task #8
[x] Simple task #9
[ ] Simple task #10
```

````python
task_dict = {}

class Task():
    def __init__(self, task):
        self.task = task
        
    def finish(self):
        new_dict = task_dict
        task_dict[Task(task)] = '[x]'
                 
    def __repr__(self):
        return str(self.task)
        
class ToDoList(Task):
    def __init__(self, list_name):
        self.list_name = list_name
        
    def add_task(self, task):
        if i % 2:
            pass
        else:
            task_dict[Task(task)] = '[ ]'
       

    def print_list(self):
        print(self.list_name)
        print('#'*3)
        for key, value in task_dict.items():
            print(value, key)
````

### Sprawdź swoje rozwiązanie wykonując kod poniżej

````python
# Weryfikacja poprawności rozwiązania
mylist = ToDoList('My sample ToDoList')
for i in range(1, 11):
    task = Task(f'Simple task #{i}')
    if i % 2:
        task.finish()
    mylist.add_task(task)
mylist.print_list()
````

## Wykonanie zapytania
Wykonaj zapytanie, które zwróci numery kont osób o nazwisku 'Nowak'. Użyj relacji między tabelami customers i accounts.

````python
connection = sqlite3.connect(r'./database.sqlite3')
cursor = connection.cursor()

cursor.execute("SELECT a.number \
                FROM accounts a inner join customers c on a.owner = c.id \
                WHERE c.last_name = 'Nowak'")

for x in cursor.fetchall():
    print(x)
````
